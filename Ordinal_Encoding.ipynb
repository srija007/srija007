```python
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/campusx-official/100-days-of-machine-learning/main/day26-ordinal-encoding/customer.csv")
```

There are 4 categocal columns. We need to convert those columns into numecial data. Gender column has no order but other 3 columns contains the order of the categorical data. So as we knew to encoding the data we need to apply ordinal encoder to the input columns and for the output columns label encoder.

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.iloc[:,2:4]
y = df.purchased
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.20)
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
Or = OrdinalEncoder(categories=[['Poor', 'Average', 'Good'], ['School', 'UG', 'PG']])
```

```python
Or.fit(X_train)
```

```python
X_train_encoded = Or.transform(X_train)
```
